```
import pandas as pd
import seaborn as sns
import matplotlib as plt

```

```
print(sns.get_dataset_names())
```

```
dataset = sns.load_dataset('diamonds')
```

```
dataset.describe
```

```
sns.scatterplot(data=dataset, x ='carat', y='price')
```

```
sns.scatterplot(data=dataset, x='carat', y='price', linewidth=0, alpha=0.3, hue='cut')
```

```
dataset.head()
```

```
dataset['depth_xtable'] = dataset['depth']*dataset['table']
dataset.head()
```

```
sns.scatterplot(data=dataset, x='carat', y='depth_xtable', linewidth=0.3, alpha=0.3)
```

```
dataset['volume'] = dataset['x'] * dataset['y'] * dataset['z']
dataset.head()
```

```
sns.scatterplot(data=dataset, x='carat', y='volume', linewidth=0.3, alpha=0.3, hue='price')
```

```
sample = dataset.sample(frac=0.05, random_state=1) #subsampling but random_state ensures you get the same one every time it generates
sample.head()
```

```
sns.pairplot(data=sample)
```

```
sns.boxplot(data=dataset, x='carat')
```

```
sns.boxplot(data=dataset, x='carat', y='clarity')
```

```
dataset.head()
```

```
pv = dataset.pivot_table(index='clarity', columns=['cut'], values='price')
sns.heatmap(pv)
```
